```python
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt
dataset = load_dataset("yelp_review_full")
```

```python
train_data = dataset["train"]
test_data = dataset["test"]
train_data = pd.DataFrame(train_data)
test_data = pd.DataFrame(test_data)
```

```python
combined_data = pd.concat([train_data, test_data], axis=0)
```

```python
plt.hist(combined_data["label"],bins=10)
plt.title("Distribution of Yelp Ratings")
plt.xlabel("Score (from 0 to 4)")
plt.ylabel("Frequency")
plt.savefig("yelp_distribution.png")
plt.show()
```

```python
drug_reviews = load_dataset("lewtun/drug-reviews")
```

```python
drug_reviews_train = drug_reviews["train"]
drug_reviews_test = drug_reviews["test"]
```

```python
drug_reviews_train = pd.DataFrame(drug_reviews_train)
drug_reviews_test = pd.DataFrame(drug_reviews_test)
```

```python
combined_drug_reviews = pd.concat([drug_reviews_train, drug_reviews_test], axis=0)
```

```python
review_drug = combined_drug_reviews["review"]
```

```python
review_drug
```

```python
average_length_chars = review_drug.apply(len).mean()
print(f"Average length of a review in the drug reviews dataset: {average_length_chars} characters")

average_length_words = review_drug.apply(lambda x: len(x.split())).mean()
print(f"Average length of a review in the drug reviews dataset: {average_length_words} words")
```
