```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
tokenizer = AutoTokenizer.from_pretrained("t5_f_experiment_0")
model_0 = AutoModelForSeq2SeqLM.from_pretrained("t5_f_experiment_0")
model_1 = AutoModelForSeq2SeqLM.from_pretrained("t5_f_experiment_1")
model_2 = AutoModelForSeq2SeqLM.from_pretrained("t5_f_experiment_2")
model_3 = AutoModelForSeq2SeqLM.from_pretrained("t5_f_experiment_3")
model_4 = AutoModelForSeq2SeqLM.from_pretrained("t5_f_experiment_4")
base_model = AutoModelForSeq2SeqLM.from_pretrained("google-t5/t5-small")
```

```python
weight_model_0 = model_0.state_dict()
weight_model_1 = model_1.state_dict()
weight_model_2 = model_2.state_dict()
weight_model_3 = model_3.state_dict()
weight_model_4 = model_4.state_dict()
base_weight = base_model.state_dict()
```

```python
for name,weight in weight_model_0.items():
    print(f"Name: {name}")
    print(f"Model 0: {weight}")
```

```python
import torch
```

```python
for (name_1, weight_1), (name_2, weight_2) in zip(weight_model_0.items(), base_weight.items()):
    if torch.all(torch.eq(weight_1, weight_2)):
        print(f"Weights for {name_1} and {name_2} are the same.")
    else:
        print(f"Weights for {name_1} and {name_2} are different.")
```
